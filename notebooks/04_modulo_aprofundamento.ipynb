# Módulo de Aprofundamento - M5: Ranking Híbrido (RRF)

Este notebook executa o modelo de busca híbrida fundindo os resultados léxicos (BM25) e semânticos (KNN) com a técnica Reciprocal Rank Fusion (RRF).

```python
import sys
from pathlib import Path
sys.path.append('../src')

from retrievers import BM25Retriever, DenseKNNRetriever, HybridRetriever
import json
import pandas as pd
```

```python
# Carrega corpus
docs = []
with open('../data/corpus.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        docs.append(json.loads(line))
print(f"Documentos carregados: {len(docs)}")
```

```python
# Inicializa os retrievers de base
bm25_ret = BM25Retriever(k1=1.5, b=0.75, mode='simple')
bm25_ret.fit(docs)

knn_ret = DenseKNNRetriever(model_name='all-MiniLM-L6-v2', cache_dir='../data')
knn_ret.fit(docs)

hybrid_ret = HybridRetriever(bm25_ret, knn_ret)
```

```python
# Exemplo de busca híbrida (RRF)
query = 'automated testing of flutter mobile applications'
results = hybrid_ret.search(query, k=5)

for rank, (doc_id, score, doc) in enumerate(results, 1):
    print(f"{rank}. [{score:.4f}] {doc['title']} (ID: {doc_id})")
```
