# Recuperador Semântico KNN Denso

Este notebook executa o modelo de busca semântica KNN usando embeddings pré-treinados (`all-MiniLM-L6-v2`) sobre a base de dados de artigos científicos.

```python
import sys
from pathlib import Path
sys.path.append('../src')

from retrievers import DenseKNNRetriever
import json
import pandas as pd
```

```python
# Carrega corpus
docs = []
with open('../data/corpus.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        docs.append(json.loads(line))
print(f"Documentos carregados: {len(docs)}")
```

```python
# Inicializa o retriever semântico KNN
retriever = DenseKNNRetriever(model_name='all-MiniLM-L6-v2', cache_dir='../data')
retriever.fit(docs) # Carrega do cache ou calcula
```

```python
# Exemplo de busca
query = 'automated testing of flutter mobile applications'
results = retriever.search(query, k=5)

for rank, (idx, score, doc) in enumerate(results, 1):
    print(f"{rank}. [{score:.4f}] {doc['title']}")
    print(f"   Abstract: {doc['abstract'][:150]}...")
```
